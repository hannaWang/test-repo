關鍵字補充：依特色主題抓景點關鍵字

(1)主城市與繁華街道

```python
import requests
from bs4 import BeautifulSoup
res = requests.get('http://www.welcome2japan.hk/location/interests/cities.html')
res.encoding = 'utf-8'
soup=BeautifulSoup(res.text)

for sec in soup.select('.paragraph tbody')[6:9:2]:#除人氣旅遊景點、日本傳統遊覽點，其他前面已都抓了
    for i in sec.select('tr td a'):
        print u' '.join(i.text.split())
```
