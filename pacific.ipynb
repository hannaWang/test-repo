```python
import requests
from bs4 import BeautifulSoup
payload={
'search_chk':'1',
'GSEARCH_MGRP_FG':'0',
'TVL_AREA':'0010',
'AMRNK':'1',
'TVL_DAREA':'0010',
'TVL_SAREA':'',
'FROM_CD':'',
'AIR_PT':'',
'SEL_TKT_BDL':'2016/06/01',
'SEL_TKT_EDL':'2016/08/30',
't_days':'',
'bodjet':'',
'CARR_CD':'',
'ikeyword':'',
'AllowJoin':'1',
'hITN_AREA':'',
'hITN_NATN':'',
'hITN_CITY':'',
'TWFG':'N',
'frm':'searcheng'}

res = requests.post('http://cowell.pac-group.net/eWeb/GO/L_GO_List_ASP.asp', data=payload)
soup=BeautifulSoup(res.text)
List=[]
for ent in soup.select('.go-contant-table tr')[1: ]:
    tourList = {}
    tourList['agency']='pacific'
    tourList['title'] = ent.select('td')[2].text 
    url='http://cowell.pac-group.net/'+ent.select('td a')[0]['href'] 
    tourList['url'] = url
    tourList['prodNo1'] = url.split('&')[0].split('MGRUP_CD=')[1]
    tourList['prodNo2'] = url.split('&')[1].split('GRUP_CD=')[1]
    
    res2 = requests.get(url)
    soup2 = BeautifulSoup(res2.text)
    tour={}
    for ent2 in soup2.select('.Q'):
        tour["day"+str(ent2.select('.go-everyday-td01')[0].text.split()[1])]=ent2.select('.go-everyday-td02')[0].text
    
    tourList['tour']=tour   
    List.append(tourList)
    
```

```python
import json
print json.dumps(List, encoding='UTF-8', ensure_ascii=False)
```

```python
with open('E:/pacific_okinawa.json' , 'w') as f:
    json.dump(List, f)
```
