```python
import requests
from bs4 import BeautifulSoup
res = requests.get('http://www.lifetour.com.tw/Oversea/SelectOversea.aspx?FROM_CD=&PORT_CD=&CITY_CD=0001&price=&SDATE=2016/05/26&EDATE=2016/08/26&order=1&after=1&keyword=')
soup = BeautifulSoup(res.text)
List=[]
for ent in soup.select('#gr_SearchItem'):
        tourList = {}
        tourList['agency']='wufu'
        tourList['prodNo']= "".join(ent.select('td')[2].text.strip().split()).split("-")[0]
        tourList['title'] = "".join(ent.select('td')[2].text.strip().split()).split("-")[1]
        url = 'http://www.lifetour.com.tw/Oversea/' +ent.select('a')[0]['href'] 
        tourList['url'] = 'http://www.lifetour.com.tw/Oversea/' +ent.select('a')[0]['href'] 
        res2 = requests.get(url)
        soup2 = BeautifulSoup(res2.text)
        #print prodNo, title, url      
        
        d=1
        tour={}
        for ent2 in soup2.select('#group_wrap .group_day_Title'):
            tour['day'+str(d)] =ent2.select('h2')[0].text
            d=d+1 
            
        
        tourList['tour']=tour
        List.append(tourList)
        
        import json
        print json.dumps(List, encoding='UTF-8', ensure_ascii=False)
```
