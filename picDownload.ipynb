Browse by Areas (Eng)

```python
import requests
from bs4 import BeautifulSoup
res = requests.get('http://www.jnto.go.jp/eng/location/destinations/index.html')
#print res.text
soup=BeautifulSoup(res.text)
for ent in soup.select('#main'):
    section = ent.select('h3') #日本八大區
    for i in section:        
        print i.text                
```
